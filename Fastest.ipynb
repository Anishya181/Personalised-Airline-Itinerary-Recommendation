```python
import pandas as pd
dataIn = pd.read_csv('data_full_anonym.csv')
dataIn.head()
```

```python
total_individuals = dataIn['individual'].nunique()
total_individuals
```

```python
dataIn['rank'] = dataIn.groupby('individual')['totalTripDurationMinutes'].rank(method='first')
dataIn.head()
```

## Top-1 Accuracy

```python
top1 = dataIn[(dataIn['choice'] == 1) & (dataIn['rank'] == 1)]
top1.head()
```

```python
top1_count = top1['individual'].nunique()
top1_count
```

```python
print("Top-1 accuracy: ", (top1_count/total_individuals))
```

## Top-5 Accuracy

```python
top5_data = dataIn[(dataIn['choice'] == 1) & (dataIn['rank'] <= 5)]
top5_data.head()
```

```python
top5_count = top5_data['individual'].nunique()
top5_count
```

```python
print("Top-5 accuracy: ", (top5_count/total_individuals))
```
